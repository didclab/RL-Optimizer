```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

job_size = 40802190232 #bytes
job_size_mb = (8*job_size)/1000000


ods_results = pd.read_csv("./results/ods_download_results.csv")

rclone_results = pd.read_csv("./results/rclone_download_results.csv",names=["job_size", "total_seconds", "mbThroughput","concurrency", "parallelism","pipe", "credId", "destRemote"])
rclone_results['mbThroughput'] = job_size_mb/rclone_results['total_seconds']


print(rclone_results)
print(ods_results)
```

```python
print(rclone_results.columns)
print(ods_results.columns)

print(rclone_results['mbThroughput'])
plt.plot(rclone_results['mbThroughput'],  label='RClone Download')
plt.plot(ods_results['throughput (Mbps)'], label='ODS Download')
plt.legend()
plt.title("Http Download Throughput")
plt.ylabel('Throughput (Mbps)')
plt.xlabel('Runs')

plt.savefig("http_throughput_graph.png")
```
